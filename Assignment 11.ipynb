```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
plt.style.use('ggplot')
```

```python
df = pd.read_csv('Data/Jackson Hole - Wyoming.csv')
#Get the first month of data
df = df.iloc[0:15]
```

```python
df.head()
```

```python
#Convert dates to correct format
df['Date'] =  pd.to_datetime(df['Date'])
#dates = mpl.dates.date2num(df['Date'])
x = df['Date']
```

```python
#Process the data by removing units
df['24 hr New Snow'] = df['24 hr New Snow'].str.rstrip('cm')
df['24 hr New Snow'] = pd.to_numeric(df['24 hr New Snow'])
y = df['24 hr New Snow'].iloc[0:15]
y.head()
```

```python
#plot
plt.plot(x, y)
```

```python
#Label x and y axis
plt.ylabel("Height(CM)")
plt.xlabel("Date")
#plot
plt.plot(x, y)
```

```python
#Label x and y axis
plt.ylabel("Height(CM)")
plt.xlabel("Date")
#Add a title
plt.title("Snowfall From January to February in Jackson Hole, Wyoming")
plt.plot(x, y)
```

```python
#Label x and y axis with proper names
plt.ylabel("Height(CM)")
plt.xlabel("Date")
#Add a title
plt.title("Snowfall From January to February in Jackson Hole, Wyoming")
#Remove Gridlines to make the graph look better
plt.grid(b=None)
plt.plot(x, y)
```

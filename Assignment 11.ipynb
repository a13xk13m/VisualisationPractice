```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
plt.style.use('ggplot')
```

```python
df = pd.read_csv('Data/Jackson Hole - Wyoming.csv')
#Get the first month of data
df = df.iloc[0:15]
```

```python
df.head()
```

```python
#Convert dates to correct format
df['Date'] =  pd.to_datetime(df['Date'])
#dates = mpl.dates.date2num(df['Date'])
x = df['Date']
```

```python
#Process the data by removing units
df['24 hr New Snow'] = df['24 hr New Snow'].str.rstrip('cm')
df['24 hr New Snow'] = pd.to_numeric(df['24 hr New Snow'])
y = df['24 hr New Snow'].iloc[0:15]
y.head()
```

```python
#plot
plt.plot(x, y)
```

```python
#Label x and y axis
plt.ylabel("Height(CM)")
plt.xlabel("Date")
#plot
plt.plot(x, y)
```

```python
#Label x and y axis
plt.ylabel("Height(CM)")
plt.xlabel("Date")
#Add a title
plt.title("Snowfall From January to February in Jackson Hole, Wyoming")
plt.plot(x, y)
```

```python
#Label x and y axis with proper names
plt.ylabel("Height(CM)")
plt.xlabel("Date")
#Add a title
plt.title("Snowfall From January to February in Jackson Hole, Wyoming")
#Remove Gridlines to make the graph look better
plt.grid(b=None)
plt.plot(x, y)
```

```python
#Label x and y axis with proper names
plt.ylabel("Height(CM)")
plt.xlabel("Date")
#Add a title
plt.title("Snowfall From January to February in Jackson Hole, Wyoming")
#Remove Gridlines to make the graph look better
plt.grid(b=None)
#Extend y limits to look better
plt.ylim(bottom=0, top=60)
plt.plot(x, y)
```

```python
#Label x and y axis with proper names
plt.ylabel("Height(CM)")
plt.xlabel("Date")
#Add a title
plt.title("Snowfall From January to February in Jackson Hole, Wyoming")
#Remove Gridlines to make the graph look better
plt.grid(b=None)
#Extend y limits to look better
plt.ylim(bottom=0, top=60)
#change the background color to make it easier to read and more fitting for a white background
plt.rcParams['axes.facecolor'] = 'white'
plt.plot(x, y)
```

```python
#Label x and y axis with proper names
plt.ylabel("Height(CM)")
plt.xlabel("Date")
#Add a title
plt.title("Snowfall From January to February in Jackson Hole, Wyoming")
#Remove Gridlines to make the graph look better
plt.grid(b=None)
#Extend y limits to look better
plt.ylim(bottom=0, top=60)
#change the background color to make it easier to read and more fitting for a white background
plt.rcParams['axes.facecolor'] = 'white'
#Add margin to make the graph easier to read
plt.title("Snowfall From January to February In Jackson Hole, Wyoming, 2017", y=1.05)
plt.plot(x, y)
```

```python
#Label x and y axis with proper names
plt.ylabel("Height(CM)")
plt.xlabel("Date")
#Add a title
plt.title("Snowfall From January to February in Jackson Hole, Wyoming")
#Remove Gridlines to make the graph look better
plt.grid(b=None)
#Extend y limits to look better
plt.ylim(bottom=0, top=60)
#change the background color to make it easier to read and more fitting for a white background
plt.rcParams['axes.facecolor'] = 'white'
#Add margin to make the graph easier to read
plt.title("Snowfall From January to February In Jackson Hole, Wyoming, 2017", y=1.05)
#Add a margin to the bottom to make it a little bit cleaner
plt.xlabel("Date", labelpad=10)
plt.plot(x, y)
```
